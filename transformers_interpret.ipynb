```python
from transformers_interpret import SequenceClassificationExplainer
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch
model_name = "roberta-base"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

model.load_state_dict(torch.load('/nfs/turbo/coe-vvh/ljr/Censorship/models/20230218-14:15:37_binary_0.1_8_1e-05_adamw.pt'))
model.eval()

# With both the model and tokenizer initialized we are now able to get explanations on an example text.

cls_explainer = SequenceClassificationExplainer(
    model,
    tokenizer)
```

```python
import pandas as pd
df = pd.read_csv('/nfs/turbo/coe-vvh/ljr/Censorship/data/dataset_test_p_en.csv', index_col=0, engine='python')
```

```python
tmp_df = df.sample(n=10, replace=False)
print(tmp_df['censored'].values, tmp_df['full_text_proc'].values)
for _, row in tmp_df.iterrows():
    word_attributions = cls_explainer(row['full_text_proc'])
    print(word_attributions)
    print(cls_explainer.visualize())
```

```python
tmp_df = df.sample(n=500, replace=False, random_state=0)
# print(tmp_df['censored'].values, tmp_df['full_text_proc'].values)
from tqdm import tqdm
from collections import Counter, defaultdict
tok_dict_0, tok_dict_1 = defaultdict(int), defaultdict(int)
tok_per_dict_0, tok_per_dict_1 = defaultdict(int), defaultdict(int)
for _, row in tqdm(tmp_df.iterrows()):
    word_attributions = cls_explainer(row['full_text_proc'])
    if cls_explainer.pred_class.item() == 0:
        for attrib in word_attributions:
            tok_dict_0[attrib[0]] += attrib[1]
            tok_per_dict_0[attrib[0]] += 1
    elif cls_explainer.pred_class.item() == 1:
        for attrib in word_attributions:
            tok_dict_1[attrib[0]] += attrib[1]
            tok_per_dict_1[attrib[0]] += 1
#     print(word_attributions)
#     print(cls_explainer.visualize())
```

```python
# get total importance
sorted(tok_dict_0.items(), key=lambda x: x[1], reverse=True)
```

```python
sorted(tok_dict_1.items(), key=lambda x: x[1], reverse=True)
```

```python
avg_dict_0, avg_dict_1 = {}, {}
for key, val in tok_dict_0.items():
    avg_dict_0[key] = val/tok_per_dict_0[key]
for key, val in tok_dict_1.items():
    avg_dict_1[key] = val/tok_per_dict_1[key]
```
